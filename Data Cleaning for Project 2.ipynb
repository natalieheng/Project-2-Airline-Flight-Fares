# Data Cleaning

```python
#rename variables
flight.rename(columns = {'Depreture  Airport' : 'Departure Airport'}, inplace = True)
flight.rename(columns = {'Depreture Time' : 'Departure Time'}, inplace = True)
flight.rename(columns = {'Ticket prize(Doller)' : 'Ticket price(Dollar)'}, inplace = True)
flight.rename(columns = {'Travel  Time' : 'Travel Time'}, inplace = True)
flight.rename(columns = {'Number of  Stoppage' : 'Number of Stoppage'}, inplace = True)
flight.rename(columns = {'Number of  Stoppage ' : 'Number of Stoppage(s)'}, inplace = True)
flight.rename(columns = {'2nd StoppageWaiting Time' : '2nd Stoppage Waiting Time'}, inplace = True)
flight.rename(columns = {'1st Stoppage Waiting Hour' : '1st Stoppage Waiting Time'}, inplace = True)
```

```python
#checking data types for above df
flight.dtypes
```

```python
#changing to string
flight[['Airline name ', 'Departure Airport', '1st Stoppage', '2nd Stoppage', '3rd Stoppage', 'Destination Airport']] = flight[['Airline name ', 'Departure Airport', '1st Stoppage', '2nd Stoppage', '3rd Stoppage', 'Destination Airport']].astype('string')
```

```python
#change travel time to hour and min
#flight['Travel Time'] 
flight.dtypes
```

```python
#replacing irrelevant values in this column to nan first 
flight['Ticket price(Dollar)'].replace('Alaska', np.nan, inplace = True)
flight['Ticket price(Dollar)'].replace('$', np.nan, inplace = True)
```

```python
#change to float
#flight['Ticket prize(Dollar)'] = flight['Ticket prize(Dollar)'].astype(float)
flight['Ticket price(Dollar)'] = flight['Ticket price(Dollar)'].astype(float).round(2)
```

```python
mean_price = np.mean(flight['Ticket price(Dollar)'])
#fill the nan values in same column with average price of ticket (based on the data set)
flight['Ticket price(Dollar)'] = flight['Ticket price(Dollar)'].fillna(mean_price)
```

```python
flight
```

```python
#eliminate unnecessary info in 'Arrival Date' column
flight['Arrival Date'].replace('Arrives: Mon, Apr 10', ' Apr 10', inplace = True)
flight['Arrival Date'].replace('Budget Friendly', ' Apr 10', inplace = True)
flight['Arrival Date'].replace('Shortest Duration', ' Apr 10', inplace = True)
flight['Arrival Date'].replace('Cheapest', ' Apr 10', inplace = True)
flight['Arrival Date'].replace('Free cancellation within 50 hours of booking', ' Apr 10', inplace = True)
```

```python
#convert arrival date
#flight['Arrival Date'] = pd.Series(flight['Arrival Date'])
flight['Arrival Date'] = pd.to_datetime(flight['Arrival Date'], format = ' %b %d')
```

```python
#extracting just the month and day without the year
flight['Arrival Date'] = flight['Arrival Date'].dt.strftime('%m-%d')
```

```python
#flight['Ticket price(Dollar)'].replace('Alaska', np.nan, inplace = True)
flight['Travel Time'].replace('Save $39!', np.nan, inplace = True)
```

```python
flight['Travel Time']
```

```python
def convert_string_to_time(string):
    if isinstance(string, str):
        # Check if the value is a string
        hours = pd.to_numeric(string.split('h')[0])
        minutes = pd.to_numeric(string.split(' ')[1].split('m')[0])

        # Concatenate hours, minutes, and seconds into a string
        time_str = f'{hours}:{minutes}:00'
        return time_str
    else:
        return string

flight['Travel Time'] = flight['Travel Time'].apply(convert_string_to_time)
```

```python
flight.head(5)
```

```python
#removing the word 'stop' from number of stoppage 
flight['Number of Stoppage(s)'] = flight['Number of Stoppage(s)'].str.replace('Stops', '')
flight['Number of Stoppage(s)'] = flight['Number of Stoppage(s)'].str.replace('Stop', '')
```

```python
flight.head(5)
```

```python
flight['Number of Stoppage(s)'].replace('Nonstop', np.nan, inplace = True)
flight['Number of Stoppage(s)'].replace('EXPRESS DEAL', np.nan, inplace = True)
```

```python
flight['Number of Stoppage(s)'] = pd.to_numeric(flight['Number of Stoppage(s)'])
```

```python
mean_stop = np.mean(flight['Number of Stoppage(s)'])
flight['Number of Stoppage(s)'] = flight['Number of Stoppage(s)'].fillna(mean_stop)
flight['Number of Stoppage(s)'] = flight['Number of Stoppage(s)'].astype(int)
```

```python
flight.head(5)
```

```python
flight['Departure Time'].replace('Sign in', np.nan, inplace = True)
```

```python
#converting departure time data type
flight['Departure Time'] = flight['Departure Time'].astype('datetime64[ns]')
```

```python
flight['Arrival Time'].replace('Delta', np.nan, inplace = True)
```

```python
#converting arrival time data type 
flight['Arrival Time'] = flight['Arrival Time'].astype('datetime64[ns]')
```
